```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 1000

# Randomly generate factor values for 1000 students
academic_performance = np.random.choice(['Poor', 'Average', 'Good'], n)
bullied = np.random.choice([True, False], n)
has_close_friends = np.random.choice([True, False], n)
homesick_level = np.random.randint(1, 6, n)
mess_food_rating = np.random.randint(1, 6, n)
sports_participation = np.random.choice([True, False], n)
social_activities = np.random.randint(0, 6, n)
study_hours = np.random.randint(0, 9, n)
screen_time = np.random.randint(1, 11, n)
sleep_hours = np.random.randint(3, 11, n)

df = pd.DataFrame({
    'sleep_hours': sleep_hours,
    'academic_performance': academic_performance,
    'bullied': bullied,
    'has_close_friends': has_close_friends,
    'homesick_level': homesick_level,
    'mess_food_rating': mess_food_rating,
    'sports_participation': sports_participation,
    'social_activities': social_activities,
    'study_hours': study_hours,
    'screen_time': screen_time
})

def assign_condition(row):
    sleep = row['sleep_hours']
    academic = row['academic_performance']
    bullied = row['bullied']
    friends = row['has_close_friends']
    homesick = row['homesick_level']
    mess = row['mess_food_rating']
    sports = row['sports_participation']
    social = row['social_activities']
    study = row['study_hours']
    screen = row['screen_time']

    # Depression
    if (sleep <= 5 and screen >= 7) or (bullied and not friends) or (academic == 'Poor' and social <= 2):
        return 'Depression'
    # Anxiety
    elif (homesick >= 4 and bullied) or (academic == 'Poor' and sleep <= 6) or (screen >= 8 and social <= 2):
        return 'Anxiety'
    # Stress
    elif (study >= 7) or (sleep <= 5 and academic == 'Poor') or (homesick >= 4 and study >= 6):
        return 'Stress'
    # ADD/ADHD
    elif (study <= 2 and screen >= 8 and academic == 'Poor') or (social >= 4 and academic == 'Poor'):
        return 'ADHD'
    # PTSD
    elif bullied and sleep <= 5 and homesick >= 4:
        return 'PTSD'
    # OCD
    elif (study >= 7 and social <= 1) or (academic == 'Good' and mess <= 2 and study >= 6):
        return 'OCD'
    # Bipolar Disorder
    elif (sleep <= 4 or sleep >= 9) and (social >= 4 or sports) and screen >= 8:
        return 'Bipolar Disorder'
    # Eating Disorder
    elif mess <= 2 and sleep <= 5 and homesick >= 4:
        return 'Eating Disorder'
    # Adjustment Disorder
    elif homesick >= 4 and academic == 'Average' and 5 <= sleep <= 7:
        return 'Adjustment Disorder'
    else:
        return 'Normal'

df['mental_health_condition'] = df.apply(assign_condition, axis=1)
df.to_csv('boarding_school_mental_health_1000.csv', index=False)
```
